```python
# Import modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

# generate random dots in x-y plane
number_of_dots = 1000
x = np.random.rand(number_of_dots,1) * 2 - 1  
y = np.random.rand(number_of_dots,1) * 2 - 1
plt.plot(x, y, 'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random dots')

# overplot a circle
xc = [math.sin(ang/2./3.14) for ang in range(360)] 
yc = [math.cos(ang/2./3.14) for ang in range(360)]
plt.plot(xc,yc,'b')
plt.show()
```

```python
# compute distances to origin
dis2 = np.array(x**2 + y**2)
id1 = np.where(dis2 <= 1)
id2 = np.where(dis2 > 1)

plt.plot(x[id1], y[id1],'ro',x[id2],y[id2],'bo')
plt.title('$\pi$ = '+str(x[id1].size*4.0/x.size))
plt.show()
```
